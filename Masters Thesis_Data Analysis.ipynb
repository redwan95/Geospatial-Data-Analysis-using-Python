```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Data: Year by year % of area according to land cover class
area_percentage_data = {
    'LC_Class': ['Waterbody', 'Vegetation', 'Developed, Low Intensity', 'Developed, Medium Intensity', 'Developed, High Intensity', 'Barren Land'],
    '2000': [3.15, 4.31, 10.49, 47.94, 32.92, 1.19],
    '2007': [3.04, 4.02, 10.14, 48.25, 33.41, 1.14],
    '2013': [2.98, 3.82, 9.77, 48.46, 33.89, 1.08],
    '2016': [2.9182, 3.774506, 9.684369, 48.50177, 33.99685, 1.124315],
    '2022': [2.86, 3.71, 9.47, 48.63, 34.25, 1.08]
}

# Data: Year by year mean land surface temperature according to land cover class
temperature_data = {
    'LC_Class': ['Waterbody', 'Vegetation', 'Developed, Low Intensity', 'Developed, Medium Intensity', 'Developed, High Intensity', 'Barren Land'],
    '2000': [30.27, 32.10, 32.77, 34.01, 34.51, 32.85],
    '2007': [33.52, 35.49, 36.76, 38.73, 39.28, 36.02],
    '2013': [28.05, 29.57, 30.38, 31.11, 31.02, 30.18],
    '2016': [26.18, 27.51, 28.78, 29.98, 30.05, 28.16],
    '2022': [33.46, 36.24, 38.19, 39.84, 40.03, 37.62]
}

# Convert data into pandas DataFrames
df_area = pd.DataFrame(area_percentage_data)
df_temp = pd.DataFrame(temperature_data)

# Extracting numerical data for correlation and regression
numeric_area = df_area.set_index('LC_Class').T.reset_index(drop=True)
numeric_temp = df_temp.set_index('LC_Class').T.reset_index(drop=True)

# Calculate correlation matrix between area and temperature
correlation_matrix = numeric_area.corrwith(numeric_temp)

# Plotting the correlation graph
plt.figure(figsize=(8, 6))
sns.barplot(x=correlation_matrix.index, y=correlation_matrix.values)
plt.title('Correlation between Land Cover Area and Mean Temperature')
plt.xlabel('Land Cover Class')
plt.ylabel('Correlation')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Performing regression analysis and plotting regression graph
plt.figure(figsize=(8, 6))
for lc_class in numeric_area.columns:
    slope, intercept, r_value, p_value, std_err = linregress(numeric_area[lc_class], numeric_temp[lc_class])
    plt.plot(numeric_area[lc_class], intercept + slope * numeric_area[lc_class], label=f'{lc_class}: R^2 = {r_value**2:.2f}')
plt.title('Regression: Area % vs. Mean Temperature')
plt.xlabel('Area %')
plt.ylabel('Mean Temperature')
plt.legend()
plt.tight_layout()
plt.show()
```
